```python
df0['FECHAS'] = pd.to_datetime(df0.apply(lambda x: x['FECHA'], 1), dayfirst=True)
df0 = df0.drop(['FECHA'], 1)
```

```python
#Columna mes
df0['MESES']=df0['FECHAS'].dt.month
df0.head()
```

```python
#Columna de años
df0['AÑOS']=df0['FECHAS'].dt.year
df0.head()
```

```python
#Dandole nombre a cada mes
df0_ENE=df0[df0.MESES==1]
df0_FEB=df0[df0.MESES==2]
df0_MAR=df0[df0.MESES==3]
df0_ABR=df0[df0.MESES==4]
df0_MAY=df0[df0.MESES==5]
df0_JUN=df0[df0.MESES==6]
df0_JUL=df0[df0.MESES==7]
df0_AGO=df0[df0.MESES==8]
df0_SEP=df0[df0.MESES==9]
df0_OCT=df0[df0.MESES==10]
df0_NOV=df0[df0.MESES==11]
df0_DIC=df0[df0.MESES==12]
```

```python
#¿Cómo le podrás determinar cuáles son los meses más lluviosos?
```

```python
#Calculamos la suma de las lluvias de cada mes y mostramos cada una en pantalla 
SUMP_ENE=df0_ENE.PRECIP.sum()
print("La suma de Enero = ",SUMP_ENE)

SUMP_FEB=df0_FEB.PRECIP.sum()
print("La suma de Febrero = ",SUMP_FEB)

SUMP_MAR=df0_MAR.PRECIP.sum()
print("La suma de Marzo = ",SUMP_MAR)

SUMP_ABR=df0_ABR.PRECIP.sum()
print("La suma de Abril = ",SUMP_ABR)

SUMP_MAY=df0_MAY.PRECIP.sum()
print("La suma de Mayo = ",SUMP_MAY)

SUMP_JUN=df0_JUN.PRECIP.sum()
print("La suma de Junio = ",SUMP_JUN)

SUMP_JUL=df0_JUL.PRECIP.sum()
print("La suma de Julio = ",SUMP_JUL)

SUMP_AGO=df0_AGO.PRECIP.sum()
print("La suma de Agosto = ",SUMP_AGO)

SUMP_SEP=df0_SEP.PRECIP.sum()
print("La suma de Septiembre = ",SUMP_SEP)

SUMP_OCT=df0_OCT.PRECIP.sum()
print("La suma de Octubre = ",SUMP_OCT)

SUMP_NOV=df0_NOV.PRECIP.sum()
print("La suma de Noviembre = ",SUMP_NOV)

SUMP_DIC=df0_DIC.PRECIP.sum()
print("La suma de Diciembre = ",SUMP_DIC)
```

```python
#Creando gráfica para la suma de las lluvias en cada mes
Meses = ('ENE', 'FEB', 'MAR', 'ABR', 'MAY', 'JUN', 'JUL','AGO','SEP','OCT','NOV','DIC')
ypos = np.arange(len(Meses))
SumaP= [SUMP_ENE,SUMP_FEB,SUMP_MAR,SUMP_ABR,SUMP_MAY,SUMP_JUN,SUMP_JUL,SUMP_AGO,SUMP_SEP,SUMP_OCT,SUMP_NOV,SUMP_DIC]
 
plt.bar(ypos, SumaP, align='center', alpha=0.7)
plt.xticks(ypos, Meses)
plt.ylabel('Precipitación acumulada (mm)')
plt.title('Suma de las lluvias por mes (1963-2011)')
plt.show()
```

```python
#¿Cuáles años han tenido inviernos fríos?
```

```python
INVTEMP= pd.DataFrame(df0.loc[df0['MESES'].isin([12,1,2])])
INVTEMP.head()
```

```python
#Arreglo para cada uno
TINVMAXAÑO = [INVTEMP[INVTEMP.AÑOS==(start + i)].TMAX.mean() for i in range(0, 49)]
TINVMINAÑO = [INVTEMP[INVTEMP.AÑOS==(start + i)].TMIN.mean() for i in range(0,49)]
```

```python
d1 = {'AÑOS': AÑOS, 'Tempmaxiviernoaño':TINVMAXAÑO ,'Tempmininviernoaño':TINVMINAÑO }
df1 = pd.DataFrame(data=d1)
df1
```

```python
#Creando gráfica para temperatura promedio mínima y máxima en inviernos por año

N = np.arange(len(fecha)) 

plt.bar(N,TINVMAXAÑO, width=0.5, align='center', alpha=0.7,color='g')
plt.xticks(N, AÑOS, size = 'small', color = 'k', rotation = 90)

plt.bar(N, TINVMINAÑO, width=0.5, align='center', alpha=0.7,color='y')
plt.xticks(N, AÑOS, size = 'small', color = 'k', rotation = 90)

plt.ylabel('Temperatura (°C)')
plt.title('Temperatura máxima y mínima promedio en invierno')
plt.gca().legend(('TMAX','TMIN'),frameon=False,loc='upper left')
plt.savefig('#Temperaturas mínimas',plt=2000)
plt.show()
```

```python
#identifica la temperatura mínima
min=df1.Tempmininviernoaño.idxmin()
df1.iloc[[min]]
```

```python
#la mínima de la mínima
df1.nsmallest(5,'Tempmininviernoaño', keep='first')
```
